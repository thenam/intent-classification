```python
import pandas as pd
```

```python
tran_df = pd.read_json('../data/processed/train.jsonl', lines=True)
val_df = pd.read_json('../data/processed/validation.jsonl', lines=True)
```

```python
len(tran_df['label'])
```

```python
len(tran_df['text'])
```

```python
from collections import Counter
import matplotlib.pyplot as plt
```

```python
label_train_counts = Counter(tran_df['label'])
plt.figure(figsize=(8, 5))
plt.bar(label_train_counts.keys(), label_train_counts.values())
plt.xlabel("Label")
plt.ylabel("Số lượng")
plt.title("Phân bổ câu theo nhãn (tập train)")
plt.show()
```

```python
label_val_counts = Counter(val_df['label'])
plt.figure(figsize=(8, 5))
plt.bar(label_val_counts.keys(), label_val_counts.values())
plt.xlabel("Label")
plt.ylabel("Số lượng")
plt.title("Phân bổ câu theo nhãn (tập validation)")
plt.show()
```
